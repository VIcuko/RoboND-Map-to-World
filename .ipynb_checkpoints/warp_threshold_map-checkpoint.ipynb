## Here we are going to combine the threshold and the transorming methods from the previous repos

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Uncomment the next line for use in a Jupyter notebook
#%matplotlib inline
import numpy as np
import cv2
```

```python
def perspect_transform(img, src, dst):

    # Get transform matrix using cv2.getPerspectivTransform()
    M = cv2.getPerspectiveTransform(src, dst)

    # Warp image using cv2.warpPerspective()
    # keep same size as input image
    warped = cv2.warpPerspective(img, M, (img.shape[1], img.shape[0]))
    
    # Return the result
    return warped
```

```python
def color_thresh(img, rgb_thresh=(0, 0, 0)):
    ####### TODO 
    # Create an empty array the same size in x and y as the image 
        # but just a single channel
    nrows, ncols, ndepth = img.shape
    filtered_image = np.zeros_like(img[:,:,0])

    # Apply the thresholds for RGB and 
        # assign 1's where threshold was exceeded        
    filtering_conditions = (img[:,:,0] > rgb_thresh[0]) & (img[:,:,1] > rgb_thresh[1]) & (img[:,:,2] > rgb_thresh[2])
    
    filtered_image[filtering_conditions]=1
        
    return filtered_image
```

```python
# Read in the same sample image as before
image = mpimg.imread('img2017-03-26-20-25-10-378.jpg')
plt.imshow(image)
plt.show()
```

```python
# Now we apply a threshold to the image

filtered_image = color_thresh(image, rgb_thresh=(160,160,0))
plt.imshow(filtered_image, cmap="gray")
plt.show()
```

```python
#Image colouring to blue (in case we want the map to be a different colour and everything apart from the floor white)
converted_image = convert_to_colour(filtered_image,[0,0,255],[0,0,0])
plt.imshow(converted_image)
plt.show()
```

## Now we transform the image into map

```python
# Define source and destination points
transformed_size = 10
offset= 6

source = np.float32([[15,141], [118,96], [199,96], [300,141]])
destination = np.float32([[image.shape[1]/2-transformed_size/2,image.shape[0]-offset], 
                          [image.shape[1]/2-transformed_size/2,image.shape[0]-transformed_size-offset], 
                          [image.shape[1]/2+transformed_size/2,image.shape[0]-transformed_size-offset], 
                          [image.shape[1]/2+transformed_size/2,image.shape[0]-offset]])      

warped = perspect_transform(converted_image, source, destination)
plt.imshow(warped)
plt.show()
```

```python
#Now we plot the coloured area into a matplotlib graph
ypos, xpos, colour = warped.nonzero()
plt.plot(xpos, ypos, '.')
plt.xlim(0, 320)
plt.ylim(0, 160)
plt.show()
```

```python
#therefore the function will be:
def rover_coords(binary_img):
    # Extract xpos and ypos pixel positions from binary_img and
    ypos, xpos, colour = binary_img.nonzero()
    # Convert xpos and ypos to rover-centric coordinates
    x_pixel = np.copy(-(ypos - binary_img.shape[0])).astype(np.float)
    y_pixel = np.copy(-(xpos - binary_img.shape[1]/2)).astype(np.float)
    
    return x_pixel, y_pixel
```

```python
xcords,ycords = rover_coords(warped)
plt.plot(xcords,ycords, '.')
plt.show()
```
